### 1) What is the average distance traveled by trips with a maximum of 2 passengers?

Getting the averange for every number of passengers.

```python
distance_plot = big_data.groupby("passenger_count")["trip_distance"].mean().plot(kind="bar", 
                                                                                 linewidth=2, 
                                                                                 color="lightblue", 
                                                                                 figsize=(10, 8), 
                                                                                 fontsize=14)
distance_plot.set_xlabel("Número de passageiros", fontdict={'fontsize':16}) 
distance_plot.set_ylabel("Distância média", fontdict={'fontsize':16}) 
distance_plot.set_title("Distância média por número de passageiros", pad=20, fontdict={'fontsize':18})
```

> ### Solution 2: I'll consider trips with 0 passengers because it can be considered in some situations like cost calc. The distance averange for trips with 0, 1 or 2 passanger is:

```python
print(f'Average distance : {big_data.query("passenger_count <= 2")["trip_distance"].mean()}')
```

### 2) Which are the 3 biggest vendors based on the total amount of money raised?

Preparing data to take vendors names

Calculating and ordering the data

```python
amount_by_vendor = big_data.groupby("vendor_id")["total_amount"].sum().reset_index()
amount_by_vendor = amount_by_vendor.sort_values(by = ["total_amount"], ascending=[False], ignore_index=True)
print(amount_by_vendor.round(5))
```

Join data to have vendors' names

```python
vendors_data = pd.read_csv(CSV_VENDORS)
amount_by_vendor = amount_by_vendor.merge(vendors_data, on="vendor_id", how="left")[["vendor_id", "name", "total_amount"]]
```

```python
print(amount_by_vendor.round(5))

```

Generating a general chart with all data

```python
amount_by_vendor_plot = amount_by_vendor.plot(kind="pie", y = "total_amount",
                                              labels=amount_by_vendor["name"],
                                              legend = False, autopct="%1.1f%%",
                                              figsize=(10, 6), 
                                              explode=(0, 0, 0, 0.3),
                                              shadow=True,
                                              startangle=0)
amount_by_vendor_plot.set_title("Proportions about every vendor", pad=30, fontdict={'fontsize':16})
```

> ### Solution: These area the top 3 biggest vendors

```python
top_vendors = amount_by_vendor.iloc[:3].round(3)
print(top_vendors)
```

```python
top_vendors_plot = top_vendors.plot(kind="pie", y = "total_amount",
                                    labels=top_vendors["name"],
                                    legend=False,
                                    figsize=(10, 6), 
                                    autopct="%1.1f%%",
                                    explode=(0.1, 0, 0),
                                    shadow=True,
                                    startangle=0)
top_vendors_plot.set_title("Proportions about top 3 vendor", pad=30, fontdict={'fontsize':16})
```

### 3) Make a histogram of the monthly distribution over 4 years of rides paid with cash.

Adjusting date types

```python
big_data["pickup_datetime"] = pd.to_datetime(big_data["pickup_datetime"])
big_data["dropoff_datetime"] = pd.to_datetime(big_data["dropoff_datetime"])
```

Data Evalating

```python
big_data.head(10)
```

Get used payment types

```python
big_data["payment_type"].unique()
```

Calculate monthly distribution for cash payments

```python
cash_only = big_data.query("payment_type in ('Cash', 'CASH')")
trips_monthly_distribution = cash_only["pickup_datetime"].groupby(cash_only.pickup_datetime.dt.to_period("M")).agg('count')
cash_monthly_distribution = cash_only.groupby(cash_only.pickup_datetime.dt.to_period("M"))["total_amount"].agg("sum")
```

Checking some data just to be aware

```python
trips_monthly_distribution.head()
```

```python
cash_monthly_distribution.head()
```

```python
# Valor médio pago em dinheiro 
cash_only.groupby(cash_only.pickup_datetime.dt.to_period("M"))["total_amount"].agg("mean")
```

Recreating as Dataframes

```python
trips_monthly_distribution = pd.DataFrame({"monthly":trips_monthly_distribution.index, "trips":trips_monthly_distribution.values})
cash_monthly_distribution = pd.DataFrame({"monthly":cash_monthly_distribution.index, "cash":cash_monthly_distribution.values})
```

```python
print(cash_monthly_distribution)
```

> ### Solution: Histogram of the monthly distribution over 4 years of rides paid with cash

```python
trip_distribution_plot = trips_monthly_distribution.plot(kind="bar",
                                                         x="monthly",
                                                         width=1.0,
                                                         linewidth=2, 
                                                         color="lightblue", 
                                                         figsize=(10, 8), 
                                                         fontsize=14, 
                                                         rot=90)
trip_distribution_plot.set_xlabel("Year-Month", fontdict={'fontsize':16}) 
trip_distribution_plot.set_ylabel("Number of trips", fontdict={'fontsize':16}) 
trip_distribution_plot.set_title("Trips distribution by month", pad=20, fontdict={'fontsize':18})
```

The histogram below is just to confirm that the monetary value has the same behavior as the number of trips

```python
cash_distribution_plot = cash_monthly_distribution.plot(kind="bar",
                                                        x="monthly",
                                                        width=1.0,
                                                        linewidth=2, 
                                                        color="lightblue", 
                                                        figsize=(10, 8), 
                                                        fontsize=14, 
                                                        rot=90)
cash_distribution_plot.set_xlabel("Year-Month", fontdict={'fontsize':16}) 
cash_distribution_plot.set_ylabel("Cash Total", fontdict={'fontsize':16}) 
cash_distribution_plot.set_title("Cash distribution by month", pad=20, fontdict={'fontsize':18})
```

### 4) Make a time series chart computing the number of tips each day for the last 3 months of 2012.

Data preparation

```python
last_months_2012 = big_data[big_data["pickup_datetime"].dt.year == 2012] #["pickup_datetime"]
```

Data visualization just for quick analysis

```python
last_months_2012.head()
```

Adjusting and calculating data by day

```python
last_months_2012 = last_months_2012['pickup_datetime'].groupby(last_months_2012.pickup_datetime.dt.to_period("D")).agg('count')
last_months_2012 = pd.DataFrame({"daily_2012":last_months_2012.index, "trips":last_months_2012.values})
```

```python
last_months_2012.head()
```

Preparing data to be ploted

```python
last_months_2012["daily_2012"] = last_months_2012["daily_2012"].dt.to_timestamp('s').dt.strftime('%Y-%m-%d %H:%M:%S.000')
last_months_2012["daily_2012"] = pd.to_datetime(last_months_2012["daily_2012"])
last_months_2012 = last_months_2012.set_index("daily_2012")
last_months_2012 = last_months_2012.last("3M")
```

```python
last_months_2012.head(300)
```

> ### Solution: Time series chart computing the number of tips each day for the last 3 months of 2012.

```python
last_months_2012_plot = last_months_2012.plot(figsize=(18, 8),
                                              fontsize=14)
last_months_2012_plot.set_xlabel("Day", fontdict={'fontsize':16}) 
last_months_2012_plot.set_ylabel("Total trips", fontdict={'fontsize':16}) 
last_months_2012_plot.set_title("Total of trips by day in last 3 months of 2012", pad=20, fontdict={'fontsize':18})
```

### 5) Make a latitude and longitude map view of pickups and dropoffs in the year 2010

Load map data of NYC

```python
nyc_street_map = gpd.read_file(NYC_MAP)
```

Defines the coordinate system to be used

```python
nyc_shape = nyc_street_map.to_crs({'init': 'epsg:4326'})
```

Show NYC shape map

```python
nyc_shape.plot()
```

Show coordinates bounds

```python
nyc_shape.total_bounds
```

Prepare data to be plotted

```python
crs = {'init': 'epsg:4326'} #{"init":"espc:4326"} #(latitude, longitude)

small_data = big_data[["pickup_longitude", "pickup_latitude", "dropoff_longitude", "dropoff_latitude"]].copy()
small_data["action"] = 1

small_data.head()
```

```python
pickups_data_geometry = [Point(coordinates) for coordinates in zip(small_data["pickup_longitude"], small_data["pickup_latitude"])]
geo_small_data = gpd.GeoDataFrame(small_data, crs=crs, geometry=pickups_data_geometry)
```

> ### Solution: latitude and longitude map view of pickups and dropoffs in the year 2010.

#### Taxi pickups map

```python
nyc_figure, nyc_axes = plt.subplots(figsize=(15,15))
nyc_street_map.plot(ax=nyc_axes, alpha=0.4, color="grey")
geo_small_data.plot(column="action", ax=nyc_axes, markersize=10, alpha=0.5, legend=True)
plt.title("Taxi pickups in NYC", fontsize=16, fontweight="bold")
plt.xlim(-74.256, -73.701)
plt.ylim(40.497, 40.916)
plt.show()
```

#### Taxi pickups map

```python
dropoff_data_geometry = [Point(coordinates) for coordinates in zip(small_data["dropoff_longitude"], small_data["dropoff_latitude"])]
geo_small_data = gpd.GeoDataFrame(small_data, crs=crs, geometry=dropoff_data_geometry)
```

```python
nyc_figure, nyc_axes = plt.subplots(figsize=(15,15))
nyc_street_map.plot(ax=nyc_axes, alpha=0.4, color="grey")
geo_small_data.plot(column="action", ax=nyc_axes, markersize=10, alpha=0.5, legend=True)
plt.title("Taxi dropoff in NYC", fontsize=16, fontweight="bold")
plt.xlim(-74.256, -73.701)
plt.ylim(40.497, 40.916)
plt.show()
```
